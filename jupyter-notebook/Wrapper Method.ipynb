```python
#df_neg = df[df < 0]
#df
df = df[(df['age']>=0) & (df['gender']>=0)  & (df['height']>=0) & (df['weight']>=0) & (df['ap_hi']>=0) & (df['ap_lo']>=0) & (df['cholesterol']>=0) & (df['gluc']>=0)  & (df['smoke']>=0) & (df['alco']>=0) & (df['active']>=0) & (df['cardio']>=0)]
df
```

```python
df["age_year"] = round(df["age"] / 365)
df.head()
```

```python
cols1 = df["weight"]
cols2 = df["height"] / 100
df["bmi"] = (cols1) / (cols2)**2

df.head()
```

```python
# class count
class_count_0, class_count_1 = df['cardio'].value_counts()

# Separate class
class_0 = df[df['cardio'] == 0]
class_1 = df[df['cardio'] == 1]# print the shape of the class
print('class 0:', class_0.shape)
print('class 1:', class_1.shape)
```
